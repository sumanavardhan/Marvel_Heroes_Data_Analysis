4.Displaying Barplots

Function for barplot(superpowers)

```python
def  create_bar(col):
    plt.figure(figsize = (10,6))
    temp_df = marvel.groupby("Alignment")[col].mean().round(decimals=2).reset_index()
    ax = sns.barplot(
        data = temp_df,
        x = "Alignment",
        y = col,
        palette = ["#941A00", "#439429", "#1E38AD"]
    )
    plt.title(f"{col} with respect to Alignment", fontsize =  15, weight = "bold");
    for i in ax.containers:
        ax.bar_label(i,)
    plt.show()
```

```python
cols = ['Intelligence', 'Strength', 'Speed', 'Durability', 'Power', 'Combat']
for col in cols:
    create_bar(col)
```

5.Showing the distribution of Total (all the powers combined) for the good and the evil.

```python
#Creating two filters to find good and bad characters.
filterGood=marvel['Alignment'].isin(['good'])
filterBad=marvel['Alignment'].isin(['bad'])

#Creating a figure for 2 plots
f,(ax1,ax2)=plt.subplots(2,1,figsize=(7,7), sharex=True, sharey=True)

#Plot the data and show the distribution
sns.distplot(marvel[filterGood]['Total'],ax=ax1)
ax1.set_ylabel("Good")
sns.distplot(marvel[filterBad]['Total'],ax=ax2)
ax2.set_ylabel("Bad")
```

6.Character Allignment Piechart

```python
labels = marvel["Alignment"].value_counts().index
sizes = marvel["Alignment"].value_counts().values
color = ["green","red","blue"]
explode =[0,0,0]
plt.subplots(figsize =(10,10))
plt.pie(sizes, explode = explode ,labels=labels, colors = color,autopct ="%1.1f%%")
plt.title("Percentages of Characters Alignment's")
plt.legend()
```
